```python
def finding(ransomNote,target,arr):
    for i in range(0,len(ransomNote)):
        # print(f"{ransomNote[i]}=={target}")
        if ransomNote[i]==target and i not in arr:
            # print(i)
            return i
    return -1

def canConstruct(ransomNote,magazine):
    ransomNote_arr=[]
    for i in magazine:
        if i in ransomNote:
            j=finding(ransomNote,i,ransomNote_arr)
            if j!=-1:
                # print(f"{j}==={ransomNote}=={i}")
                ransomNote_arr.append(j)
                if len(ransomNote_arr)==len(ransomNote):
                    return True
        # print(f"{ransomNote_arr}==={ransomNote}")

    return False
            
  
```
